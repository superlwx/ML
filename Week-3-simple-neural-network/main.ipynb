```python
import numpy as np
import matplotlib.pyplot as plt
from lr_utils import load_dataset
from planar_utils import sigmoid
```

```python
def G(x):
    return np.maximum(x, 0.01*x)
def gra_G(x):
    result=np.zeros(x.shape)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i][j]>0:
                result[i][j]=1
            else:
                result[i][j]=0.01
    return result
X,Y,X_test,Y_test,classs=load_dataset()
X=X.reshape((X.shape[0],-1)).T
X_test=X_test.reshape((X_test.shape[0],-1))
w1=np.random.randn(4,X.shape[0])*0.01
print(w1)
b1=np.zeros((4,1))
w2=np.random.randn(1,4)*0.01
b2=np.zeros((1,1))
alpha=0.01
m=X.shape[0]
for i in range(2000):
    z1=np.dot(w1,X)+b1#(4,2)*(2,400)+(4,1)=(4,400)
    a1=G(z1)#(4,400)
    z2=np.dot(w2,a1)+b2#(1,4)*(4,400)+(1,1)=(1,400)
    a2=sigmoid(z2)#(1,400)
    dz2=a2-Y#(1,400)
    dw2=(1/m)*np.dot(dz2,a1.T)#(1,400)*(400,4)=(1,4)
    db2=(1/m)*np.sum(dz2,axis=1).reshape(1,1)#(1,1)
    dz1=np.multiply(np.dot(w2.T,dz2),gra_G(z1))#(4,1)*(1,400)*(4,400)
    dw1=(1/m)*np.dot(dz1,X.T)#(4,2)
    db1=(1/m)*np.sum(dz1,axis=1).reshape(4,1)#(4,1)
    w1=w1-alpha*dw1
    w2=w2-alpha*dw2
    b1=b1-alpha*db1
    b2=b2-alpha*b2
#X_test,Y_test=load_planar_dataset()
z1_test=np.dot(w1,X)+b1
a1_test=G(z1_test)
z2_test=np.dot(w2,a1_test)+b2
Y_prediction=G(z2_test)
print(Y_prediction)
for i in range(Y_prediction.shape[1]):
    Y_prediction[0,i]=1 if Y_prediction[0,i]>0.5 else 0
print("测试集准确性："  , format(100 - np.mean(np.abs(Y_prediction - Y)) * 100) ,"%")
```
